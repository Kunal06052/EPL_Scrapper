```python
import os
import re
import csv
import time
import logging
import random
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

def setup_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--disable-blink-features=AutomationControlled")
    options.add_argument("--start-maximized")
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    return driver

def human_delay(a=1, b=3):
    time.sleep(round(random.uniform(a, b), 2))

def extract_match_details(driver, match_url):
    logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")
    try:
        driver.get(match_url)
        human_delay(2, 5)


        try:
            accept_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//button[contains(text(), "Accept")]'))
            )
            accept_button.click()
            human_delay(1, 2)
        except:
            pass

        

        
            # Wait and extract the match details
        for _ in range(3):
            try:
                stats_tab = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH, '//*[@id="mainContent"]/div/section[2]/div[2]/div/div[1]/div/div/ul/li[3]'))
                )
                stats_tab.click()
                break  # Exit loop if click is successful
            except Exception as e:
                time.sleep(2)  # Retry after a delay


        
        # Handling cookie pop-ups
        # try:
        #     accept_button = WebDriverWait(driver, 10).until(
        #         EC.element_to_be_clickable((By.XPATH, '//button[contains(text(), "Accept")]'))
        #     )
        #     accept_button.click()
        #     human_delay(1, 2)
        # except:
        #     pass
        
        # Extract match details
        try:
            match_details_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="mainContent"]/div/section[2]/div[2]/section/div[1]/div[1]'))
            )
            match_text = match_details_element.text.strip()
        except:
            logging.warning(f"Match details not found for {match_url}")
            return None
        
        # Extract teams
        try:
            h_team_name = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div/section[2]/div[2]/section/div[1]/div[2]/div[1]/div[1]/div[1]/a[2]').text.strip()
            a_team_name = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div/section[2]/div[2]/section/div[1]/div[2]/div[1]/div[3]/div[1]/a[2]').text.strip()
        except:
            logging.warning(f"Team names not found for {match_url}")
            return None
        
        # Extract full-time and half-time scores
        full_time_result = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div/section[2]/div[2]/section/div[1]/div[2]/div[1]/div[2]/div[1]').text.strip()
        h_time_result = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div/section[2]/div[2]/section/div[1]/div[2]/div[1]/div[2]/div[2]').text.strip()
        
        # Parse scores
        try:
            FTHG, FTAG = map(int, full_time_result.split(" - "))
            FTR = "H" if FTHG > FTAG else "A" if FTHG < FTAG else "D"
        except:
            FTHG, FTAG, FTR = None, None, None
        
        try:
            match = re.search(r"(\d+)-(\d+)", h_time_result)
            HTHG, HTAG = map(int, match.groups()) if match else (None, None)
            HTR = "H" if HTHG > HTAG else "A" if HTHG < HTAG else "D"
        except:
            HTHG, HTAG, HTR = None, None, None
        
        # Process match details
        match_details_lines = match_text.split("\n")
        date, time_match, ground, attendance, referee = match_details_lines[0], match_details_lines[2], match_details_lines[3], match_details_lines[4].replace("Att: ", ""), match_details_lines[5].replace("Ref: ", "")
        
        return {
            "Date": date, "Time": time_match, "Ground": ground,
            "Attendance": attendance, "Referee": referee,
            "Home Team": h_team_name, "Away Team": a_team_name,
            "FTHG": FTHG, "FTAG": FTAG, "FTR": FTR,
            "HTHG": HTHG, "HTAG": HTAG, "HTR": HTR
        }
    except Exception as e:
        logging.error(f"Error extracting data for {match_url}: {e}")
        return None

def process_season_matches(driver, season_file):
    season = season_file.replace("matches_", "").replace(".csv", "").replace("_", "/")
    output_file = f"{season}_all_match_stats.csv"
    logging.info(f"Processing matches for {season}")
    
    with open(season_file, mode="r", encoding="utf-8") as file:
        reader = csv.reader(file)
        next(reader)
        match_links = [row[1] for row in reader]
    
    all_match_data = []
    for idx, match_url in enumerate(match_links):
        logging.info(f"Processing match {idx+1}/{len(match_links)}: {match_url}")
        match_data = extract_match_details(driver, match_url)
        if match_data:
            all_match_data.append(match_data)
        human_delay(5, 10)
    
    if all_match_data:
        pd.DataFrame(all_match_data).to_csv(output_file, index=False)
        logging.info(f"✅ Data saved: {output_file}")
    else:
        logging.warning(f"No data extracted for {season}")

def main():
    logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")
    driver = setup_driver()
    
    season_files = sorted(
        [f for f in os.listdir() if f.startswith("matches_") and f.endswith(".csv")],
        key=lambda x: int(x.split("_")[1].split(".")[0])
    )
    filtered_seasons = [f for f in season_files if "2009_10" <= f.replace("matches_", "").replace(".csv", "") <= "2024_25"]
    
    for season_file in filtered_seasons:
        process_season_matches(driver, season_file)
        logging.info("⏳ Taking a short break...")
        human_delay(random.int(500,1000), random.int(800, 1200))
    
    driver.quit()
    logging.info("✅ All seasons processed successfully!")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import glob
import requests
import time

# Define the function to process the link
def process_link(url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # Check for HTTP errors
        return len(response.text)  # Example: Returning length of the HTML content
    except requests.RequestException as e:
        return str(e)

# Specify folder path
folder_path = "path_to_your_folder/*.csv"

# Output file
output_file = "processed_data.csv"

# Create/Open output file and add headers if necessary
with open(output_file, "w") as f:
    f.write("filename,link,result\n")  # Modify headers as needed

# Loop through each CSV file in the folder
for file in glob.glob(folder_path):
    df = pd.read_csv(file)  # Read CSV file
    filename = file.split("/")[-1]  # Extract filename

    # Assuming links are in a column named "link"
    if "link" in df.columns:
        for link in df["link"].dropna():
            result = process_link(link)  # Apply function to the link
            with open(output_file, "a") as f:
                f.write(f"{filename},{link},{result}\n")  # Append result
            time.sleep(1)  # Pause to avoid overloading the server

print("Processing complete. Data saved in:", output_file)
```
