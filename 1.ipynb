```python
import time
import random
import logging
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from concurrent.futures import ThreadPoolExecutor
from selenium.common.exceptions import TimeoutException, NoSuchElementException

# Set up logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

BASE_URL = "https://www.premierleague.com/results"

# Function to scroll dynamically to load more matches
def human_like_scroll(driver, pause_time=2, max_scrolls=10):
    last_height = driver.execute_script("return document.body.scrollHeight")
    
    for _ in range(max_scrolls):
        driver.execute_script("window.scrollBy(0, 500);")  # Scroll down
        time.sleep(random.uniform(1, pause_time))  # Randomized delay

        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break  # Stop if no new content loads
        last_height = new_height

# Function to extract match links from the dynamically loaded divs
def extract_match_links(driver, max_scrolls=10):
    match_links = []
    
    try:
        for _ in range(max_scrolls):
            human_like_scroll(driver, pause_time=2)

            # Find all match divs
            matches = driver.find_elements(By.XPATH, '//div[contains(@class, "match-fixture__wrapper")]')

            for match in matches:
                match_id = match.get_attribute("data-matchid")
                relative_link = match.get_attribute("data-href")
                full_link = f"{BASE_URL}{relative_link}" if relative_link else None
                
                # if full_link and match_id not in [m['match_id'] for m in match_links]:
                #     # Extract basic match details
                #     teams = match.find_elements(By.XPATH, './/span[@class="match-fixture__team-name"]/span[@class="match-fixture__short-name"]')
                #     score = match.find_element(By.XPATH, './/span[@class="match-fixture__score"]').text.strip()
                #     stadium = match.find_element(By.XPATH, './/span[@class="match-fixture__stadium-name"]').text.strip()

                #     if len(teams) == 2:
                #         home_team = teams[0].text.strip()
                #         away_team = teams[1].text.strip()

                #         match_links.append({
                #             "match_id": match_id,
                #             "url": full_link,
                #             "home_team": home_team,
                #             "away_team": away_team,
                #             "score": score,
                #             "stadium": stadium
                #         })

            time.sleep(random.uniform(2, 4))

        logging.info(f"✅ Extracted {len(match_links)} match links!")
    
    except Exception as e:
        logging.error(f"❌ Error extracting match links: {e}")

    return match_links

# Function to extract match details
def extract_match_data(match):
    try:
        options = webdriver.ChromeOptions()
        options.add_argument("--headless")
        driver = webdriver.Chrome(options=options)

        driver.get(match["url"])
        human_like_scroll(driver, pause_time=3, max_scrolls=3)

        # Extract match date & time
        try:
            match_info = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="mainContent"]/div/section[2]/div[2]/section/div[1]/div[1]'))
            ).text.strip()
            match_date, match_time = match_info.split("\n")[:2]
        except:
            match_date, match_time = None, None

        # Extract full-time & half-time results
        try:
            full_time_result = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div/section[2]/div[2]/section/div[1]/div[2]/div[1]/div[2]/div[1]').text.strip()
            FTHG, FTAG = map(int, full_time_result.split(" - "))
            FTR = "H" if FTHG > FTAG else "A" if FTHG < FTAG else "D"
        except:
            FTHG, FTAG, FTR = None, None, None

        # Store extracted data
        match_data = {
            "Match ID": match["match_id"],
            "URL": match["url"],
            "Date": match_date,
            "Time": match_time,
            "Home Team": match["home_team"],
            "Away Team": match["away_team"],
            "Full-Time Score": match["score"],
            "Stadium": match["stadium"],
            "FTHG": FTHG,
            "FTAG": FTAG,
            "FTR": FTR
        }

        # Save to CSV
        df = pd.DataFrame([match_data])
        df.to_csv("match_data.csv", mode='a', header=not pd.io.common.file_exists("match_data.csv"), index=False)

        logging.info(f"✅ Data saved for Match ID: {match['match_id']}")
        driver.quit()

    except Exception as e:
        logging.error(f"❌ Error processing match {match['match_id']}: {e}")

# Function to process multiple match pages in parallel
def process_links_parallel(matches):
    with ThreadPoolExecutor(max_workers=3) as executor:
        executor.map(extract_match_data, matches)

# Main execution
if __name__ == "__main__":
    # options = webdriver.ChromeOptions()
    # options.add_argument("--headless")
    # driver = webdriver.Chrome(options=options)

    driver.get("https://www.premierleague.com/results")
    human_like_scroll(driver, pause_time=2)

    matches = extract_match_links(driver)
    driver.quit()

    process_links_parallel(matches)
```

```python
def extract_match_links(driver, container_xpath, link_xpath, max_scrolls=10):
    match_links = set()

    try:
        for _ in range(max_scrolls):
            human_like_scroll(driver, pause_time=2)
            
            # Find the parent container that holds all matches
            try:
                container = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, container_xpath)))
            except TimeoutException:
                logging.warning("⚠️ Could not find the matches container!")
                break
            
            # Extract match links from the child divs
            match_elements = container.find_elements(By.XPATH, link_xpath)
            for element in match_elements:
                link = element.get_attribute("href")
                if link and link.startswith("http"):  # Validate extracted link
                    match_links.add(link)

            # Wait for more content to load
            time.sleep(random.uniform(2, 4))

        logging.info(f"✅ Extracted {len(match_links)} match links!")
    
    except Exception as e:
        logging.error(f"❌ Error extracting match links: {e}")

    return list(match_links)
```
